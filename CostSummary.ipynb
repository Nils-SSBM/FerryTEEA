```python
import numpy as np
from matplotlib import pyplot as plt

species = (
    "Base",
    "BB",
    "OBS",
    "WS",
    "WF",
)
weight_counts = {
    "Acquisition cost": np.array([42468, 37511, 37511, 45602, 45602]),
    "Battery part of acquisition cost": np.array([1422, 3008, 2265, 327, 1585]),
    "Battery replacement cost": np.array([3555, 7520, 4216, 818, 3964]),
    "Maintenance cost": np.array([8219, 7671, 7671, 7671, 7671]),
    "Fuel cost": np.array([9174, 7370, 7370, 7370, 7370]),
    "Infrastructure cost": np.array([188, 377, 377, 1387, 1016]),

}
totCost = [65029, 63459, 59412, 63176, 67211]
y_errormin = [totCost[0] - 62840, totCost[1] - 59784, totCost[2] - 56665, totCost[3] - 60286, totCost[4] - 64411] 
y_errormax = [66886 - totCost[0], 66583 - totCost[1], 61515 - totCost[2], 64977 - totCost[3], 69155 - totCost[4]]
yerror = [y_errormin, y_errormax]
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(5)
colors = ['blue', 'green', 'blue', 'blue', 'blue']

i=0
for boolean, weight_count in weight_counts.items():

    if i == 0:
        p = ax.bar(species, weight_count, width, label=boolean,  bottom=bottom)
        bottom += weight_count
        i += 1
    elif i == 1:
        p = ax.bar(species, weight_count, width, label=boolean, color='#1f77b4', hatch='xxxxx', bottom=bottom)
        bottom += weight_count
        i += 1
    elif 1 < i < 5:
        p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
        bottom += weight_count
        i += 1
    else:
        p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom, yerr=yerror, capsize=7)
        bottom += weight_count
    
ax.set_ylabel('Total cost (SEK / day)')
ax.legend(loc="best")

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

N = 60
P_cons = 140 #kW
t_trip = 6 #minutes
t_char = np.linspace(0.25, 2.5, 100)
#t_char = [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
#P_char = [0, 1000, 100]

def calcPchar(t_char, B):
    P = (P_cons + ((P_cons * (t_trip / 60)) / (t_char / 60)) - ((0.7 * B) / (N * (t_char / 60)))) 
    return P

P_char0 = calcPchar(t_char, 0)
P_char100 = calcPchar(t_char, 100)
P_char200 = calcPchar(t_char, 200)
P_char300 = calcPchar(t_char, 300)
P_char400 = calcPchar(t_char, 400)
P_char500 = calcPchar(t_char, 500)
P_char600 = calcPchar(t_char, 600)

fig, ax = plt.subplots()
ax.plot(t_char, P_char0, c='black', label='0 kWh')
ax.plot(t_char, P_char100, c='indigo', label='100 kWh')
ax.plot(t_char, P_char200, c='purple', label='200 kWh')
ax.plot(t_char, P_char300, c='crimson', label='300 kWh')
ax.plot(t_char, P_char400, c='red', label='400 kWh')
ax.plot(t_char, P_char500, c='darkorange', label='500 kWh')
ax.plot(t_char, P_char600, c='gold', label='600 kWh')
plt.grid(visible=None, which='major', axis='both')
ax.axis([0, 2.5, 0, 1000])
ax.set_xlabel('$t_{char}$ (minutes)')
ax.set_ylabel('$P_{char}$ (kW)')

a = np.linspace(0,2.5,100)
A = (a / 2.5) * 4  #For the right axis, showing the area of the charger pads

ax2 = ax.twinx()
ax2.set_ylabel('Charger pad area (m$^2$)')  # we already handled the x-label with ax1
ax2.plot(a, A, visible=False)
ax2.axis([0, 2.5, 0, 4])
leg = ax.legend(title='Additional capacity');
```
